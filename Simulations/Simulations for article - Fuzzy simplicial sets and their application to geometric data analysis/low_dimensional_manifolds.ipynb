```python
import sys
import os
SCRIPT_DIR = os.getcwd()
PATH_CURRENT = os.path.join(SCRIPT_DIR, "../../src/")
PLOT_PATH = os.path.join(SCRIPT_DIR, "../results/Simulations for article - Fuzzy simplicial sets and their application to geometric data analysis/low_dimensional_manifolds/")
scriptPath = os.path.abspath(PATH_CURRENT)
sys.path.append(scriptPath)
```

```python
from isumap import isumap
from data_and_plots import plot_data, printtime, createMammoth, createNonUniformHemisphere, createSwissRole, createTorus, make_s_curve_with_hole, createSCurve
from time import time
```

```python
def run_isumap(k,N,distBeyondNN,normalize,tconorm,epm,distFun,datasetCreator,datasetName,plot_path=PLOT_PATH,**datasetCreator_kwargs):
    data, labels = datasetCreator(N,**datasetCreator_kwargs)

    plot_data(data,labels,title=datasetName,display=True, save=True,path=plot_path)

    t0=time()
    finalInitEmbedding, finalEmbedding, clusterLabels = isumap(data, k, 2,
        normalize = normalize, distBeyondNN=distBeyondNN, tconorm = tconorm, epm=epm, distFun=distFun,verbose=True)
    t1 = time()
    printtime("Isumap total time",t1-t0)

    title = datasetName + " N_" + str(N) + " k_" + str(k) + " beyondNN_" + str(distBeyondNN) + " normalize_" + str(normalize) + " tconorm_" + tconorm + " distFun_" + distFun + " epm_" + str(epm)

    plot_data(finalInitEmbedding,labels,title=title+" init",display=True, save=True,path=plot_path)
    plot_data(finalEmbedding,labels,title=title,display=True, save=True,path=plot_path)
    print("\nResult saved in '../results/Simulations for article - Fuzzy simplicial sets and their application to geometric data analysis/low_dimensional_manifolds/")

    # return finalInitEmbedding, finalEmbedding, clusterLabels
```

```python
k = 15
N = 5000

distBeyondNN = False
epm = True
normalize = True
tconorm = "canonical"
distFun = "canonical"
```

```python
run_isumap(k,N,distBeyondNN,normalize,tconorm,epm,distFun,createNonUniformHemisphere,datasetName="NonUniformHemisphere")
```

```python
run_isumap(k,N,distBeyondNN,normalize,tconorm,epm,distFun,createTorus,"Torus",**{'seed': 0})
```

```python
run_isumap(k,N,distBeyondNN,normalize,tconorm,epm,distFun,createSwissRole,datasetName="SwissRole")
```

```python
run_isumap(k,N,distBeyondNN,normalize,tconorm,epm,distFun,createSCurve,datasetName="s_curve")
```
